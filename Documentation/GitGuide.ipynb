# Git Guide

## Basic commands

`git pull`: pulls from the virtual version of the repo (what lives on GitHub)

`git status`: gives you the status of files that have been added to the staging area (in green) and those yet to be added (in red). 

`git add`: used to add files to the staging area. only add the files you want to commit next. 

`git commit -m "message"`: makes changes to your local repo. you need to include a message to make a commit.

`git push`: sends the changes to the local repo to GitHub

`git branch`: shows you the branches on the repo. the * tells you what branch you are in.


## Git flow

First open the Terminal:

Git -> Open Git Repository in Terminal

### Updating local repo

`git pull origin main`

use this command before you start coding to update all the files in your local repo. this will pull the work others have pushed between the last time you worked on the repo and now. make sure you pull from the main branch by adding "origin main" to the pull request. 

### Checking your branch

`git branch`

if you are not on the right branch use:

git branch branchname

example: `git branch alrg`

### Commiting and pushing changes

`git status` -> check what files have changes and that you might want to commit

`git add` -> copy/paste the name of the file you want to stage. if you want to stage multiple files, you have to add one by one. make sure that you are on the right branch. 

git status -> use this to check on the files you staged. these files should be in green. 

`git commit -m "message about the commit"` -> this will add the changes to your local repo. You need to include a message about the change you are making. 

`git push` -> this will send the changes to GitHub
