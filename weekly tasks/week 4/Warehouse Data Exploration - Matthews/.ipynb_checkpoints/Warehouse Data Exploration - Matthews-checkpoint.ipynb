# Warehouses in Ontario California

## Data

A few years, a researcher began publishing a datadase of warehouses across California which has been updated on a regular basis. A link to that source can be found [here](https://radicalresearch.shinyapps.io/WarehouseCITY/).

```python
##Import the necessary Libraries and warehouse CSV data
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

Warehouses = pd.read_csv('Data/Warehouse_Data.csv',
        dtype=
        {
            'Assessor parcel number':str,
            'Building classification':str,
        }
)

Warehouses.sample(5)
Warehouses.info()
```

## Tax Assessor
We can also bring in the shape file of the San Bernandino County Tax assessor parcels, filtered for just Ontario City. It can be found at this [link](https://open.sbcounty.gov/datasets/sbcounty::sbco-parcel-polygons/explore?filters=eyJKdXJpc2RpY3Rpb24iOlsiQ2l0eSBvZiBPbnRhcmlvIl19&location=33.982901%2C-117.581233%2C13.83). Using this, we can determine which warehouses fall in Ontario.

```python
#Import the necesary Ontario parcel shapefile
San_Bern_Parcels = gpd.read_file('Data/Ontario Parcels/SBCo_Parcel_Polygons.shp')

San_Bern_Parcels.rename(columns = {'ParcelNumb':'Assessor parcel number'}, inplace = True)

#merge the warehouses and parcel shapefile on the San Bernandino County Tax Assessor parcel Number
Warehouse_Parcel=Warehouses.merge(San_Bern_Parcels,on="Assessor parcel number")

Warehouse_Parcel.sample(5)
Warehouse_Parcel.info()
```

This data set contains interesting information on the types and the age of warehouses. The code below looks at the data available in just the city of Ontario California. The first is a pie chart of the types of warehouses present in Ontario. The second is a historagram showing the dates of construction for all Ontario warehouses which are still on the tax roll.

```python
# Creating pie chart of warehouse types in Ontario
Warehouse_Parcel_count = Warehouse_Parcel['Building classification'].value_counts()
ax = Warehouse_Parcel_count.plot(kind='pie',title='Percent of earehouses in Ontario California by type',autopct='%1.0f%%')
ax.get_yaxis().set_visible(False)
```

```python
# Creating bar graph of warehouse construction in Ontario
Warehouse_Parcel_year = Warehouse_Parcel['Year built'].value_counts()

fig, ax = plt.subplots()
Warehouse_Parcel_year = Warehouse_Parcel['Year built'].value_counts().sort_index()
Warehouse_Parcel_year.plot(ax=ax, kind='bar',figsize=(10, 8))
```

## Parcel Visualization
Mergining this data set with the shapefile of tax assessors parcel, can help use get a visual distribution of the warehouses.

```python
Warehouse_Parcel.info()
Warehouse_Parcel_Geo = gpd.GeoDataFrame(Warehouse_Parcel, geometry='geometry', crs="EPSG:4326")

import contextily as ctx

ax = Warehouse_Parcel_Geo.plot(figsize=(12,10),
                column='Building classification',
                 legend=True)

ctx.add_basemap(ax, crs=4326)
```
